```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_input = pd.read_csv("input.dat", sep="\t", names=["u", "v", "th"])

z = np.linspace(0, 2000, len(df_input))

fig, (ax_uv, ax_th) = plt.subplots(1, 2, figsize=(8, 6))
ax_uv.plot(df_input["u"], z, label="u")
ax_uv.plot(df_input["v"], z, label="v")
ax_th.plot(df_input["th"], z, label="theta")
```
